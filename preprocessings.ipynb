- Géométrie des secteurs de bureau de vote de la Métropole de Lyon : data.grandlyon
- Résultats des élections européennes et législatives : datagouv

```python
import pandas as pd
```

### Extraire les résultats électoraux des bureaux de vote appartenant à la métropole de Lyon

```python
communes_metrop_lyon = pd.read_csv(
    filepath_or_buffer="./Data source/communes-de-la-metropole-de-lyon--a-partir-de-2024.csv",
    sep=";"
)
communes_metrop_lyon
```

```python
legis_all_dataset = pd.read_csv(
    filepath_or_buffer="./Data source/legis_resultats-definitifs-par-bureau-de-vote.csv",
    sep=";")
legis_all_dataset
```

```python
legis_metrop_lyon = pd.merge(
    left = legis_all_dataset,
    left_on=["Code commune", "Libellé commune"],
    right = communes_metrop_lyon[["nom", "insee"]], 
    right_on=["insee", "nom"])
legis_metrop_lyon
```

```python
europ_all_dataset = pd.read_csv(
    filepath_or_buffer="./Data source/europ_resultats-definitifs-par-bureau-de-vote.csv",
    sep=";")
europ_all_dataset
```

```python
europ_metrop_lyon = pd.merge(
    left = europ_all_dataset,
    left_on=["Code commune", "Libellé commune"],
    right = communes_metrop_lyon[["nom", "insee"]], 
    right_on=["insee", "nom"])
europ_metrop_lyon
```
